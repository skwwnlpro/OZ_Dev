#### 리팩토링 과제

Day4에서 만든 "붕어빵 자동 판매 시스템"은 절차적 코드로 구현되어있습니다.

Day5에서 배운 함수, 클래스 개념을 적용시켜 객체지향적으로 리팩토링을 진행해주세요


```python
import logging
import os

# 객체 생성
logger = logging.getLogger()

# 기본 로그 레벨을 INFO로 설정
# DEBUG, INFO, WARNING, ERROR, CRITICAL
logger.setLevel(logging.INFO)

# 로그 메시지의 형식을 정의
formatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')

# log를 console에 출력
stream_handler = logging.StreamHandler()
stream_handler.setFormatter(formatter)
logger.addHandler(stream_handler)

class BungeoppangShop:
    def __init__(self):
        print("Hello FishMan")
        self.stock = {
            "팥붕어빵": 10, 
            "슈크림붕어빵": 8,
            "초코붕어빵": 5
        }
        self.prices = {
            "팥붕어빵" : 1000,
            "슈크림붕어빵" : 1200,
            "초코붕어빵" : 1500,
        }

        self.sales = {
            "팥붕어빵": 0,
            "슈크림붕어빵": 0,
            "초코붕어빵": 0,
        }

    # print로 stock 정보를 출력하고 None을 Return합니다.
    def check_stock(self) -> None:
        print(self.stock)

    def process_order(self) -> None:
        try:
            bread_type = input(" # 붕어빵의 종류를 고르세요 : ")
            bread_count = int(input(" # 양의 정수인 개수를 입력하세요 : "))
            # if bread_type in self.stock:
            if self.stock[bread_type] >= bread_count:
                self.stock[bread_type] -= bread_count
                self.sales[bread_type] += bread_count
                print(f" # 판매 완료\n{self.stock}")
                print("=======================================\n")
            else: 
                print(f"요구하신 {bread_type}, {bread_count}는 재고가 부족합니다.\n다른 상품을 입력하세요.")
                print(self.stock)
            # else:
            #     # 이미 if 절로 예외를 처리해서 강제로 raise 메소드로 발생
            #     raise KeyError("!! 정확한 빵의 종류를 입력하시오 !!\n")
        except KeyError:
            print("=======================================\n")
            print("!! 정확한 빵의 종류를 입력하시오 !!\n")
        except ValueError:
            print("=======================================\n")
            print("!! 양의 정수인 개수를 정확히 입력하시오 !!\n")

    # print로 stock,additional_stock 정보를 출력하고 None을 Return합니다.
    def admin_mode(self) -> None:
    # isinstance(additional_stock, int)는 add~_stock이 int Class 인스턴스 확인
    # additional_stock > 0 은 0보다 큰지 음수 확인
        try:
            bread_type = input(" # 붕어빵 종류를 고르세요 : ")
            additional_stock = int(input(" # 양의 정수인 개수를 입력하세요 : "))
            # if bread_type in self.stock or isinstance(additional_stock, int) or additional_stock > 0:
            self.stock[bread_type] += additional_stock
            print(f" # 추가된 항목\n ### {bread_type} : {additional_stock}개\n")

        except KeyError:
            logger.warning("!! KeyError 입니다 !!")
            print("==============================\n")
            print("!! 종류를 정확히 입력하시오 !!\n")
        except ValueError:
            print("=======================================\n")
            print("!! 양의 정수인 개수를 정확히 입력하시오 !!\n")

    
    def calculate_total_sales(self):
        sum = 0
        # self.prices와 self.sales 딕셔너리를 zip으로 연결하여 순회합니다.
        # key는 self.prices의 키를 사용하고, _는 self.sales의 값을 무시합니다.
        # dictionary comprehension을 사용하여 새로운 딕셔너리 fishAmount를 생성
        # 각 키에 대해 가격(self.prices[key])과 판매량(self.sales[key])을 곱함
        fishAmount = {key: self.prices[key] * self.sales[key] for key, _ in zip(self.prices, self.sales)}
        for i in fishAmount.values():
            sum += i
        print(f"총 매출: {sum}원")

def main():
    fish = BungeoppangShop()
    powerFish = True
    
    while powerFish:
        mode = input(" # Mode 중 하나를 선택하세요. -> (Order/Admin/Exit/Clear)\n= ")
        
        if mode == "Order":
            fish.process_order()
            
        elif mode == "Admin":
            fish.admin_mode()
            
        elif mode == "Exit":
            print("# Goodbye ")
            powerFish = False
        
        elif mode == "Clear":
            os.system('cls')
            # Linux, Mac
            # os.system('clear')    
    fish.calculate_total_sales()

if __name__ == "__main__":
    main()
```

[과제 4: 주문 처리 메서드 process_order]

목표: process_order 함수를 구현하여 손님의 주문을 처리하고 재고를 업데이트합니다.

설명: bread_type과 bread_count를 매개변수로 받아, 재고를 확인하고 충분한 경우 재고를 감소시키고 판매량을 증가시킵니다. 재고가 부족하거나 잘못된 붕어빵 종류인 경우 적절한 메시지를 출력하세요.

```python

def process_order(self, bread_type, bread_count):
    try:
        if bread_type in self.stock:
            if self.stock[bread_type] >= bread_count:
                self.stock[bread_type] -= bread_count
                self.sales[bread_type] += bread_count
                print(f" # 판매 완료\n{self.stock}")
                print("==============================\n")
            else: 
                print(f"요구하신 {bread_type}, {bread_count}는 재고가 부족합니다.\n다른 상품을 입력하세요.")
                print(self.stock)
        else:
            raise
    except:
        print("==============================")
        print("!! 종류를 정확히 입력하시오 !!\n")
```

[과제 5: 관리자 모드 메서드 admin_mode]

목표: admin_mode 메서드를 구현하여 관리자 모드에서 재고를 추가할 수 있도록 합니다.

설명: bread_type과 additional_stock을 입력받아 재고를 추가합니다. 잘못된 입력 값이 들어온 경우 적절한 메시지를 출력하세요.

```python
# print로 stock,additional_stock 정보를 출력하고 None을 Return합니다.
def admin_mode(self, bread_type, additional_stock) -> None:
    # isinstance(additional_stock, int)는 add~_stock이 int Class 인스턴스 확인
    # additional_stock > 0 은 0보다 큰지 음수 확인
    try:
        if bread_type in self.stock or isinstance(additional_stock, int) or additional_stock > 0:
            self.stock[bread_type] += additional_stock
            print(f"추가된 항목 {bread_type}: {additional_stock}")

    except KeyError:
        logger.warning("!! KeyError 입니다")
        print("==============================")
        print("!! 종류를 정확히 입력하시오 !!\n")
```


[과제 6: 매출 계산 메서드 calculate_total_sales]

목표: calculate_total_sales 메서드를 구현하여 총 매출을 계산하고 출력합니다.

설명: self.sales와 self.prices를 사용하여 총 매출을 계산하고 출력하세요.

```python
def calculate_total_sales(self):
    sum = 0
    fishAmount = {key: self.prices[key] * self.sales[key] for key, _ in zip(self.prices, self.sales)}
    for i in fishAmount.values():
        sum += i
    print(f"총 매출: {sum}원")
```
