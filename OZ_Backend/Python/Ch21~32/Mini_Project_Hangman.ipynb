```python
# ê¸°ë³¸ ë¼ì´ë¸ŒëŸ¬ë¦¬ë¼ íŒŒì´ì¬ ì„¤ì¹˜ì‹œ ìë™ìœ¼ë¡œ ì„¤ì¹˜ë©ë‹ˆë‹¤!
# random ëª¨ë“ˆì€ ëœë¤í•œ ìˆ«ìë¥¼ ìƒì„±í•  ë•Œ ì‚¬ìš©í•©ë‹ˆë‹¤.
import random
import os
# ê²Œì„ì— ì‚¬ìš©ë  ë‹¨ì–´ ëª©ë¡
words = ["apple", "banana", "orange", "grape", "lemon"]

# í–‰ë§¨ ê·¸ë¦¼
hangman_pics = [
    """
     ------
     |    |
     |
     |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |    |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|\\
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|\\
     |   /
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|\\
     |   / \\
     |
    ---""",
]


class HangmanGame:
    def __init__(self, word):
        print("Game Start")
        self.word = word
        self.life = 10

    def display(self, life):
        print("========================\n")
        if life > 6:
            print(hangman_pics[0])
            print("========================\n")
        else:
            print(hangman_pics[6-life])
            print("========================\n")
            
    def play(self):
        emptyWord = ["_"] * len(self.word)
        
        while True:
            
            print(f"ë§ì¶œ ë‹¨ì–´ = {emptyWord}")
            ## Input
            inSpell = input("ì…ë ¥ ê¸€ì : ")
            
            ## Screen Clear
            os.system("cls")
            
            ## Checking & Fill 
            for spell in range(len(self.word)):
                letter = self.word[spell]         
                if letter == inSpell:
                    emptyWord[spell] = inSpell
            
            ## Success or Fail
            if inSpell in emptyWord:
                print("ë‹¨ì–´ë¥¼ ë§ì¶”ì…¨êµ°ìš”, ë” í˜ë‚´ë´ìš”.")
                print(f"ë‚¨ì€ ì‹œë„ëŠ” {self.life}ë²ˆ")
            else:
                self.life -= 1
                print(f"í‹€ë ¸ìŠµë‹ˆë‹¤. ë‚¨ì€ ì‹œë„ëŠ” {self.life}ë²ˆ")
                    
            self.display(self.life)     

            ## Done!
            if self.word == ''.join(emptyWord):
                return print(f"ì¶•í•˜í•©ë‹ˆë‹¤! ë‹¨ì–´ë¥¼ ë§ì¶”ì…¨ìŠµë‹ˆë‹¤! \në‚¨ì€ ì‹œë„ëŠ” {self.life}\n")
            
            ## GoodBye!
            if self.life == 0:
                return print("ì£½ì—ˆìŠµë‹ˆë‹¤! ì˜ê°€ì„¸ìš”!") 
            

if __name__ == "__main__":
    game = HangmanGame(random.choice(words))
    game.play()
```
